```python
df_num = df.select_dtypes("number")
df_cat = df.select_dtypes("object")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_hnum = scaler.fit_transform(df_num)
```

```python
df_hnum_s = pd.DataFrame(scaled_hnum, columns=df_num.columns)
```

```python
df_hnum_s.head()
```

```python
corr_num = df_hnum_s.corr()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(40,40))
sns.heatmap(corr_num, annot=True)
plt.show()
```

```python
from scipy.stats import pearsonr


def coef_pvalue(df, features, target):
    for feature in features:
        corr, p_value = pearsonr(df[feature], df[target])
        print(f"{feature}:\nCoef: {corr}\nP_value: {p_value}")
        print()


features = ["OverallQual", "YearBuilt", "YearRemodAdd", "TotalBsmtSF", "1stFlrSF", "GrLivArea", "FullBath", "TotRmsAbvGrd", "GarageCars", "GarageArea"]
target = "SalePrice"
coef_pvalue(corr_num, features, target)
```

```python
high_corr_num = df_hnum_s[features].corr()
```

```python
# FINDING OUTLIERS
sns.set(style="whitegrid")
sns.boxplot(df_hnum_s[features])
plt.xticks(rotation="vertical")
plt.show()
```

```python
#  CLOSER LOOK TO THOSE WITH STRONG CORRELATION TO SALEPRICE
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
sns.heatmap(high_corr_num,annot=True)
plt.show()
print("Multicollinearity might be present")
```

```python
# SCATTER PLOTS TO UNDERSTAND IF THE RELATIONSHIP IS LINEAR OR NOT
for feature in features:
    sns.regplot(x=df_hnum_s[feature], y=df_hnum_s["SalePrice"], data=df_hnum_s, line_kws={'color': 'red', 'label': "Regression Line"})
    plt.legend()
    plt.title(f"{feature} vs SalePrice")
    plt.show()
```
